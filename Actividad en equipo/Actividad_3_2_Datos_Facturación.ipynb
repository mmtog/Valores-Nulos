```python
#Importar librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_excel("Datos de Facturaci√≥n.xlsx")
data.head(5)
```

```python
#Identificar valores nulos por columna .isnull()
valores_nulos=data.isnull().sum()
valores_nulos
```

```python
#Realizamos una copia del dataframe (copy)
data1=data.copy()
```

```python
#m√©todo de sustituci√≥n de valores nulos .fillna(algo)
#Sustituir valores nulos por una clave nueva para identificar "F99999"

data1["CVE_DOC"]=data1["CVE_DOC"].fillna("F99999")
```

```python
#Corroborar valores nulos
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python
#m√©todo de sustituci√≥n de valores nulos .fillna(algo)
#Sustituir valores nulos por una leyenda que representa que no existe algun estatus
data1["STATUS"]=data1["STATUS"].fillna("SIN ESTATUS")
```

```python
#m√©todo de sustituci√≥n de valores nulos .fillna(algo)
#Sustituir valores nulos por un una leyenda que representa valores nulos
data1["CVE_VEND"]=data1["CVE_VEND"].fillna("SIN CLAVE")
```

```python
#m√©todo de sustituci√≥n de valores nulos .fillna(algo)
#Sustituir valores nulos por valores no nulos hacia adelante "forward fill" ("ffill")
#Las 3 fechas sustituye lo datos nulos por la fecha anterior
data1["FECHAELAB"]=data1["FECHAELAB"].fillna(method="ffill")
data1["FECHA_DOC"]=data1["FECHA_DOC"].fillna(method="ffill")
data1["FECHA_ENT"]=data1["FECHA_ENT"].fillna(method="ffill")
```

```python
#Corroborar valores nulos
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python
#m√©todo de sustituci√≥n de valores nulos .fillna(algo)
#Sustituir valores nulos por una leyenda que representa que no existe cancelaci√≥n en ese pedido
data1["FECHA_CANCELA"]=data1["FECHA_CANCELA"].fillna("No hay cancelaci√≥n")
```

```python
#Corroborar valores nulos
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python
#Primer m√©todo de sustituci√≥n de valores nulos .fillna(algo)
#Sustituir valores nulos con promedio o media
data1["CAN_TOT"]=data1["CAN_TOT"].fillna(round(data["CAN_TOT"].mean(),1))
```

```python
#Corroborar valores nulos
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python
#m√©todo de sustituci√≥n de valores nulos .fillna(algo)
#Sustituir valores nulos por valor numerico 0 que representa que no existe un descuento
data1["DES_TOT"]=data1["DES_TOT"].fillna(0)
data1["DES_FIN"]=data1["DES_FIN"].fillna(0)
```

```python
#Corroborar valores nulos
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python
#Corroboramos valores nulos en todo el dataframe
valores_nulos=data1.isnull().sum().sum()
valores_nulos
```
