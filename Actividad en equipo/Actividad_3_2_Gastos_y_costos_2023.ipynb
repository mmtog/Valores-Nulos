```python
#Importar librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Cargar dataframe
df = pd.read_excel("Gastos y costos 20-23.xlsx", sheet_name="2023")
df
```

```python
#Eliminar primera fila o filas inecesarias
#Se eliminan las primeras 3 filas
df=pd.read_excel("Gastos y costos 20-23.xlsx", sheet_name="2023", skiprows=3)
df
```

```python
#Eliminar columan inecesaria .drop()
#axis = 1 para columnas
#axis = 0 para fila
#Eliminamos la columna "otros" porque no existen registros
df= df.drop(["Otros"], axis=1)
df
```

```python
#Identificar valores nulos por columna .isnull()
valores_nulos=df.isnull().sum()
valores_nulos
```

```python
#Realizamos una copia del dataframe (copy)
df1=df.copy()
```

```python
#método de sustitución de valores nulos .fillna(algo)
#Sustituir por una clave nueva para identificar folio nulo. "F99999"

df1["Folio"]=df1["Folio"].fillna("F99999")
```

```python
#Corroborar valores nulos
valores_nulos=df1.isnull().sum()
valores_nulos
```

```python
#método de sustitución de valores nulos .fillna(algo)
#Sustituir valores nulos por #N/D para identificar que no hay registro
df1["TIPO GASTO"]=df1["TIPO GASTO"].fillna("#N/D")
```

```python
#Corroborar valores nulos
valores_nulos=df1.isnull().sum()
valores_nulos
```

```python
#método de sustitución de valores nulos .fillna(algo)
#Sustituir valores nulos por pagado
df1["MP"]=df1["MP"].fillna("PAGADO")
df1["FP"]=df1["FP"].fillna("PAGADO")
```

```python
#Corroborar valores nulos
valores_nulos=df1.isnull().sum()
valores_nulos
```

```python
#método de sustitución de valores nulos .fillna(algo)
#Sustituir valores nulos por sin poliza
df1["Poliza"]=df1["Poliza"].fillna("SIN POLIZA")
```

```python
#Corroborar valores nulos
valores_nulos=df1.isnull().sum()
valores_nulos
```

```python
#Corroboramos valores nulos en todo el dataframe
valores_nulos=df1.isnull().sum().sum()
valores_nulos
```

```python
df1.to_csv("Gastos y costos 20-23_Sin_Nulos.csv")
```
