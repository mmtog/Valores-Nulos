**1ERA HOJA DE EXCEL *2020***

```python
data3_2020 = pd.read_excel('Gastos y costos 20-23.xlsx', sheet_name='2020', skiprows=5)
data3_2020
```

```python
#Identificar valores nulos por columna
valores_nulos2020=data3_2020.isnull().sum()
valores_nulos2020
```

```python
#Realizamos una copia del dataframe
data3_1=data3_2020.copy()
```

```python
#Método de sustitución de valores nulos
#Sustituir valores nulos por un string en  concreto 
data3_2020["FOLIO"] =data3_2020["FOLIO"].fillna("01234567890") 
data3_2020
```

```python
#Corroboramos valores nulos
valores_nulos2020=data3_2020.isnull().sum()
valores_nulos2020
```

```python
#Método de sustitución de valores nulos
#Sustituir valores nulos por un valor numérico en  concreto 
data3_2020["GASTO"]= data3_2020["GASTO"].fillna(0) 
data3_2020
```

```python
#Corroboramos valores nulos
valores_nulos2020=data3_2020.isnull().sum()
valores_nulos2020
```

```python
data3_2020.info()
```

```python
#Método de sustitución de valores nulos
#Sustituir valores nulos con promedio o media
data3_2020["TC"]=data3_2020["TC"].fillna(round(data3_1["TC"].mean(),1))
```

```python
#Corroboramos valores nulos
valores_nulos2020=data3_2020.isnull().sum()
valores_nulos2020
```

```python
#Método de sustitución de valores nulos
#Sustituir valores nulos con promedio o media
data3_2020["IMPORTE"]=data3_2020["IMPORTE"].fillna(round(data3_1["IMPORTE"].mean(),1))
```

```python
#Corroboramos valores nulos
valores_nulos2020=data3_2020.isnull().sum()
valores_nulos2020
```

```python
#Método de sustitución de valores nulos
#Sustituir valores nulos con promedio o media
data3_2020["IVA"]=data3_2020["IVA"].fillna(round(data3_1["IVA"].mean(),1))
```

```python
#Corroboramos valores nulos
valores_nulos2020=data3_2020.isnull().sum()
valores_nulos2020
```

```python
#Método de sustitución de valores nulos
#Sustituir valores nulos por valores no nulos hacia adelante  "forward fill" ("ffill")
#Filtro por columnas
data3_2020["TIPO"] =data3_2020["TIPO"].fillna(method="ffill")  
```

```python
#Corroboramos valores nulos
valores_nulos2020=data3_2020.isnull().sum()
valores_nulos2020
```

```python
#Método de sustitución de valores nulos
#Sustituir valores nulos por un string en  concreto 
data3_2020["POLIZA"] =data3_2020["POLIZA"].fillna("NO LOS METÍ") 
data3_2020
```

```python
#Corroboramos valores nulos
valores_nulos2020=data3_2020.isnull().sum()
valores_nulos2020
```

```python
#Corroboramos valores nulos en todo el dataframe
valores_nulos2020=data3_2020.isnull().sum().sum()
valores_nulos2020
```

```python
#Convertir DataFrame a CSV
data3_2020.to_csv("Gastos y costos 2020.csv")
```
