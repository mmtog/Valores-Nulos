```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#cargar archivo csv desde COLAB
data=pd.read_excel("Detalle precios y productos fabricados 2022.xlsx")
data.head(5)
```

```python
data1= pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
data1
```

```python
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python
data2=data1.copy()
```

```python
data2["CVE_DOC"]=data2["CVE_DOC"].fillna("F11551")
data2
```

```python
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
data2["FECHA_DOC"]=data2["FECHA_DOC"].fillna("2004-12-01 00:00:00")
data2
```

```python
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
data2["NOMBRE_VENDEDOR"]=data2["NOMBRE_VENDEDOR"].fillna("Oscar Maydon")
data2
```

```python
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
data2["NOMBRE_CLIENTE"]=data2["NOMBRE_CLIENTE"].fillna("Justin Bieber")
data2
```

```python
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
data2["CANT"]=data2["CANT"].fillna(round(data2["CANT"].mean(),1))
```

```python
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
data2["CVE_ART"]=data2["CVE_ART"].fillna("TS00000")
```

```python
data2["DESCR"]=data2["DESCR"].fillna("Arriba tec pue")
```

```python
data2["PRECIO_UNITARIO"]=data2["PRECIO_UNITARIO"].fillna(round(data2["PRECIO_UNITARIO"].mean(),1))
```

```python
data2["COSTO_UNITARIO"]=data2["COSTO_UNITARIO"].fillna(round(data2["COSTO_UNITARIO"].mean(),1))
```

```python
data2["COSTO_UNITARIO_CALCULADO"]=data2["COSTO_UNITARIO_CALCULADO"].fillna(round(data2["COSTO_UNITARIO_CALCULADO"].mean(),1))
```

```python
data2["SUBTOTAL_PARTIDA"]=data2["SUBTOTAL_PARTIDA"].fillna(round(data2["SUBTOTAL_PARTIDA"].mean(),1))
```

```python
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Corroboramos valores nulos en todo el dataframe
valores_nulos=data2.isnull().sum().sum()
valores_nulos
```

```python
#Convertir DataFrame a CSV
data2.to_csv("detalleprecios_y_productos_sin_nulos.csv")
```
