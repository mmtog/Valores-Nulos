```python
#Importar las librerias pandas, numpy y matplotlib respectivamente
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Cargar desde un archivo .csv sin indice
dataframe = pd.read_excel('Gastos y costos 20-23 (1).xlsx', sheet_name= '2022')
dataframe
```

```python
#Eliminar primera fila o filas inecesarias
dataframe= pd.read_excel('Gastos y costos 20-23 (1).xlsx', sheet_name= '2022', skiprows=3)
dataframe
```

```python
dataframe.info()
```

```python
#Identificar valores nulos por columna
valores_nulos=dataframe.isnull().sum().sum()
valores_nulos
```

```python
#Identificar valores nulos por columna
valores_nulos=dataframe.isnull().sum()
valores_nulos
```

```python
#Realizamos una copia  del dataframe
dataframe1=dataframe.copy()
```

```python
dataframe.head
```

```python
#primer metodo de sustitucion de valores nulos
#sustituir valores nulos con promedio o media
dataframe1["TC"]=dataframe1["TC"].fillna(round(dataframe["TC"].mean(),1))
```

```python
#segundo metodo de sustitucion de valores nulos
#Sustituir valores nulos por un string en concreto
dataframe1["MP"]=dataframe1["MP"].fillna("PAGADO")
```

```python
#Corroboramos valores nulos
valores_nulos=dataframe1.isnull().sum()
valores_nulos
```

```python
#segundo metodo de sustitucion de valores nulos
#Sustituir valores nulos por un string en concreto
dataframe1["Poliza"]=dataframe1["Poliza"].fillna("Sin Poliza")
```

```python
#Corroboramos valores nulos
valores_nulos=dataframe1.isnull().sum()
valores_nulos
```

```python
#Tercer metodo de sustitucion de valores nulos
#Sustituir valores nulos por un valor n√∫merico en concreto 
dataframe1["Otros"]= dataframe1["Otros"].fillna(0)
dataframe1
```

```python
#Corroboramos valores nulos
valores_nulos=dataframe1.isnull().sum()
valores_nulos
```

```python
#Tercer metodo de sustitucion de valores nulos
#Sustituir valores nulos por un valor n√∫merico en concreto 
dataframe1["Folio"]= dataframe1["Folio"].fillna(000000)
dataframe1
```

```python
#Corroboramos valores nulos
valores_nulos=dataframe1.isnull().sum()
valores_nulos
```

```python
#Convertir DataFrame a CSV
dataframe1.to_csv("Gastos y costos 2022_sin_nulos.csv")
```
